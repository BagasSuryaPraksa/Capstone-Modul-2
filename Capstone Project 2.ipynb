#### **3.6.2. Generation**

Ditambahkan rentang usia berdasarkan generasi yang diperoleh dari kolom Year_Birth.
- **Post War**: Between 1928 - 1945
- **Boomers**: Between 1946 - 19664
- **Gen X**: Between 1965 - 1980
- **Millennials**: Between 1981 - 1996
- **Gen Z**: After 1996
<br>
<br>
source: https://www.beresfordresearch.com/age-range-by-generation/

```python
df["Year_Birth"].unique()
```

```python
def generation(year_birth):
    if year_birth >= 1928 and year_birth <= 1945:
        return "Post War"
    elif year_birth >= 1946 and year_birth <= 1964:
        return "Boomers"
    elif year_birth >= 1965 and year_birth <= 1980:
        return "Gen X"
    elif year_birth >= 1981 and year_birth <= 1996:
        return "Millennials"
    else:
        return "Gen Z"
```

```python
df["Generation"] = df["Year_Birth"].apply(generation)
df[["Year_Birth", "Generation"]].head()
```

```python
df["Generation"].value_counts()
```


Kita dapat menghapus pelanggan dalam generasi pascaperang karena jumlahnya sangat sedikit.

```python
df = df[~(df["Generation"]=="Post War")]
```

#### **3.6.3. Social Class**
Ditambahkan Kelas Sosial yang diperoleh dari kolom Pendapatan.
- **Lower Class**: Below $28,007
- **Lower Middle Class**: Between $28,008 and $55,000
- **Middle Clas**: Between $55,001 - $89,744
- **Upper Middle Class**: Between $89,745 - $149,131
- **Upper Class**: $149,132 or higher
<br>
<br>
source: https://www.fool.com/the-ascent/personal-finance/articles/does-your-income-make-you-upper-class-middle-class-or-lower-class/

```python
df["Income"].unique()
```

```python
def socclass(income):
    if income <= 28007:
        return "Lower Class"
    elif income >= 28008 and income <= 55000:
        return "Lower Middle Class"
    elif income >= 55001 and income <= 89774:
        return "Middle Class"
    elif income >= 89745 and income <= 149131:
        return "Upper Middle Class"
    else:
        return "Upper Class"
```

```python
df["SocClass"] = df["Income"].apply(socclass)
df[["Income", "SocClass"]].head()
```

```python
df["SocClass"].value_counts()
```

**Menunjukan rata - rata produk yang terjual berdasarkan setiap kampanye dalam bentuk Lineplot**

### **6.2. Recommendations**

**Strategi Pemasaran Berdasarkan Generasi Pelanggan**

**1. Gen X (1027 pelanggan)**

        - Kampanye Konsolidasi: 
        
            Meskipun Gen X tidak mendominasi dalam kampanye tertentu, mereka tetap signifikan dalam jumlah. Sebaiknya gunakan kampanye yang lebih luas dan serbaguna yang dapat menjangkau banyak pelanggan sekaligus.
        
        - Produk yang Direkomendasikan: 
        
        Fokus pada produk seperti anggur dan daging yang memiliki daya tarik luas, karena produk-produk ini paling banyak terjual.
        
        - Personalization: 
        
        Gunakan data pembelian sebelumnya untuk menyesuaikan penawaran produk yang sesuai dengan preferensi Gen X, misalnya melalui email marketing yang dipersonalisasi.

**2. Boomers (735 pelanggan)**

        - Kampanye Targeted: 
        
        Memperbanyak kampanye 1, 2, dan 4 karena Boomers paling banyak merespons metode ini.
        
        - Promosi Produk Premium: 
        
        Promosikan produk premium seperti anggur dan daging karena Boomers paling banyak membeli produk.

        - Loyalty Programs: 
        
        Implementasikan program loyalitas yang memberi penghargaan kepada pelanggan setia, dengan penawaran eksklusif pada produk yang sering dibeli.
        
**3. Millennials (344 pelanggan)**

        - Kampanye Digital:
        
        Manfaatkan kampanye 3, 5, dan 6 yang lebih interaktif dan berbasis media sosial, karena Millennials lebih cenderung merespons metode ini.
        
        - Produk Baru dan Berbeda:
        
        Introduksi produk baru atau produk yang lebih trendi dan ramah lingkungan, karena generasi ini sering mencari hal-hal baru.
        
        - Engagement Melalui Media Sosial: 
        
        - Aktif di platform media sosial untuk meningkatkan keterlibatan dengan generasi ini, melalui konten yang menarik dan kampanye berbasis media sosial.
        
        
**Strategi Berdasarkan Kelas Sosial**

**1. Middle Class (929 pelanggan)**

        - Kampanye Konsisten: 
        
        Pertahankan dan tingkatkan kampanye 1, 2, 4, 5, dan 6 karena kelas sosial ini paling banyak menerimanya.
        
        - Volume Purchase Promotions: 
        
        Tawarkan promosi pembelian dalam volume besar, karena Middle Class paling banyak membeli produk.

        - Produk dengan Value:
        
        Fokus pada produk dengan nilai yang baik (harga terjangkau dengan kualitas tinggi) untuk mempertahankan loyalitas pelanggan kelas ini.
        
**2.Lower Middle Class**

        - Kampanye Fokus pada Kampanye 3: 
        
        Mengarahkan kampanye 3 yang paling banyak diterima oleh kelas sosial ini.
        
        - Affordable Product Offerings: 
        
        Tawarkan produk dengan harga yang lebih terjangkau atau paket promosi yang lebih menarik.
        
        - Upselling dan Cross-selling: 
        
        Tawarkan produk tambahan yang relevan untuk meningkatkan nilai rata-rata pembelian.
        
**3. Lower Class**

        - Kampanye Khusus pada Kampanye 3 dan 6: 
        
        Fokus pada kampanye 3 dan 6 yang menunjukkan respons positif dari kelas sosial ini.
